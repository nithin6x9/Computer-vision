### Images as Signals
Images are two dimensional signals representing the visual information emerging from a scene. Each value of the independent variable of the image signal corresponds to the infromation from a single point on the scene. The visual information Correponding to a point from the scene is the intensity/color of the light ray emerging from that point. This information is usually captured with a digital camera. The Charge Coupled Devices (CCDs) from the CCD array equipped in the digital camera produce a voltage proportional to the intensity of the light ray orginating from each point on the scene. 

As traditional cameras can record only $2D$ information, we assume that a scene is two dimensional, with it's two axes representing the distance along two perpendicular directions. The camera maps the light rays from the scene area (field of view of the camera) into a rectangular $2D$ signal, where, each point $(x,y)$ on the signal corresponds to a point on the scene, and the value of the signal $f(x,y)$ at this point denotes the intensity of the light ray emerging from that point. Hence, the pair values $(x,y)$ is the independent variable (similar to the time axis in the $1D$ signals that you have studied), and the value $f(x,y)$ at that point is the signal amplitude. The above discussion applies to a monochrome image, also called a grayscale image, in which the visual information is the intensity of the light represented as a real valued number. In the case of color images, there are three values corresponding to the primary colors $R,G$ and $B$ at every $(x,y)$. Thus, the signal value at every $(x,y)$ of a color image is a $3\times 1$ vector containing the values corresponding to three primary colors. 
ImageFormation.png
As the variables $x$ and $y$ map to distances in space (of the scene), they are called spatial coordinates. Theoretically, images can be seen as a $2D$ continuous signal whose value is defined for any spatial coordinate $(x,y)$, where, $x,y\in\mathbb{R}$. But, we know that it is impractical to store and process continuous time signals using digital computers. Hence, digital cameras sample the spatial coordinates into regular discrete points, and the intensity value at each discrete point is recorded by the CCD array as shown below. In a discretized image represented as $g[m,n]$, each sample value of intensity level at any given $(m,n)$ is represented as an unsigned 8-bit integer. As an unsigned 8-bit integer can take any value between $0$ and $255$, we can can see that the intensity of light is recorded by quantizing it into 256 levels. In the case of color images, each sample value $g[m,n]$ is a $3\times 1$ vector containing three values corresponding to the primary colors.
ImageSampling.png



### Images as $2D$ Arrays
The value $g[m_0,n_0]$ of one sample of the discretized **grayscale** image at any $(m_0,n_0)$ is called a 'Pixel' (Picture Element). Thus, an image with $M$ samples along the horizontal axis, and $N$ samples along the vertical axis has a total of $MN$ pixels. This image can be seen as a $2D$ array with $M$ rows and $N$ columns as shown below.
ImageArray.png

In the case of a discretized **color** image, one sample $g[m_0,n_0]$ is a $3\times 1$ vector for any $(m_0,n_0)$. Hence, a color image with $M$ samples along the horizontal axis, and $N$ samples along the vertical axis has $MN$ pixels each for representing the primary colors Red, Green and Blue. Thus, it can be interpreted as three $MN$ sized grayscale images, with $MN$ pixels from any one of them representing a primary color. Such an interpreted grayscale image is called a color channel (Red channel, Green channel etc.). From the above discussion, it is clear that a color image can be represented as an $M\times N\times 3$ array. This representation is shown below.

ColorImage.png

As there is a difference in the convention of directions of $m$ and $n$ axes ($m$ is increasing towards right and $n$ is increasing towards top), and the traditional ordering of rows and columns (row index is increasing **downwards** and column index is increasing towards **right**), it is advantegous to change the image axes in such a way that, they align with the row and column ordering. The $m$ axis is taken to be increasing along rows and the $n$ axis taken to be increasing along columns. The graphical representation of an image with this new coordinate system is shown below.

#### Visit through each of the pixels values and set it to the highest value (255) if the present values if greater than 125
This operation is called thresholding. Here, all the pixel values above 125 are thresholded to the highest value 255. Try doing the reverse, i.e., threshold all the pixel values smaller than 125 to the lowest value 0.

```
J = np.zeros((I.shape[0],I.shape[1]));
for m in range(0,I.shape[0]):
  for n in range(0,I.shape[1]):
    if (I[m,n] > 125):
      J[m,n] = 255;
    else:
      J[m,n] = I[m,n];

cv2_imshow(J)      
```

### Read a color image and set all of it's pixel values from the Red channel to 0.
This operation will essentially kill the Red color from the image. Observe it's effect.

```
I_color = cv2.imread('/content/gdrive/MyDrive/Colab Notebooks/IPCV/Images/House.png',1)
J_color = I_color;
for m in range(0,I_color.shape[0]):
  for n in range(0,I_color.shape[1]):
    J_color[m,n,1] = 0;

cv2_imshow(J_color)
```

### Exercise 3 Image Scaling (Zoom In)
In this exercise, you have to scale (zoom in) a **grayscale** image by an integer factor $q=2$. Image scaling or magnification is similar to the signal expansion in time domain. Image scaling is also known as image super resolution. Scaling an image increases the size of the image, and hence the total number of pixels. If an image of size $M\times N$ is scaled by a factor $2$, the resulting image has a size $2M\times 2N$ with a total $4MN$ pixels. Image scaling by factor of 2 can be pictorially represented as follows. The figure on the left shows an example $4\times 4$ input image to be magnified into $8\times 8$, whereas, the figure on the right side shows an $8\times 8$ image with the pixels available from the original input image copied to the alternating pixel locations. The Red dots indicate the available pixel values from the input image. The pixel values at the locations without dots are missing, and hence needed to be interpolated.

SR1.png

There exist a wide variety of methods for filling in the missing pixel values. Here, we adopt a very simple and intuitive technique, though not very effective in terms of the quality of interpolation. This simple interpolation is carried out in two steps as shown in the following figure. 

The missing pixel values on the rows and columns with alternating red dots (whose values are available from the input) are interpolated first. Along the horizontal direction, the missing values are interpolated as the average of the available pixels from the immediate left and right positions of a missing value. Similarly, the missing values in the vertical direction are interpolated as the average of the pixels from above and below. These interpolated pixel values are shown as Green dots.

Interp.png

In the second step, the remaining missing pixels are interpolated. These pixels are interpolated as the average of the four available pixel values located diagonally around the missing value. Example input and output images are shown below.
